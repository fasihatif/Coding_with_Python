### 4 - exercise
Sort the trip, station and weather file lists. Make 3 dataframes by reading in the first csv (the first element from each sorted lists). <br><br>

```python
# Files are not being sorted correctly. Numeric sorting inside strings is not being followed.
# Example: trip_10.csv,trip_100.csv,trip_11.csv is not correctly sorted even after using sorted()

# Install natsort library that deals with natural sorting

import os
from natsort import natsorted

# Sort the trip, station and weather file lists
sorted_trip_file_list = natsorted(trip_file_list)

sorted_station_file_list = natsorted(station_file_list)

sorted_weather_file_list = natsorted(weather_file_list)
```

### 5 - exercise
Create an empty list called trip_dataframes <br>
Loop through the filenames in the sorted trip file list. Inside the loop read in the corresponding csv into a dataframe. Append this dataframe to the trip_dataframes. <br>
After the loop finished concatenate the dataframes in the trip_dataframes list into a dataframe called trip. <br><br>
Similarly create a station and a weather dataframe.

```python
# Read all files related to each category

# Trip files
trip_dataframes = []
for x in sorted_trip_file_list:
    trip_df = pd.read_csv(x)
    trip_dataframes.append(trip_df)
trip = pd.concat(trip_dataframes)

# Station files
station_dataframes = []
for y in sorted_station_file_list:
    station_df = pd.read_csv(y)
    station_dataframes.append(station_df)
station = pd.concat(station_dataframes)

# Weather files
weather_dataframes = []
for z in sorted_weather_file_list:
    weather_df = pd.read_csv(z)
    weather_dataframes.append(weather_df)
weather = pd.concat(weather_dataframes)
```
