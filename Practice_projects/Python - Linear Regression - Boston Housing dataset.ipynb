```python
#import libraries
import numpy as np
import pandas as pd

# Import Linear Regression model from Sklearn
from sklearn import linear_model
```

```python
#import the Boston Housing dataset
from sklearn.datasets import load_boston
boston = load_boston()
print(boston.keys())

# data = the independent variables also known as X values
# feature_names = the column names of the data
# target = dependable variable also known as the y value / price of the houses
```

```python
#Tranform dataset into a data frame
DFboston = pd.DataFrame(boston.data , columns = boston.feature_names)

# Add Target (Y) to the Dataframe
DFboston['Price'] = pd.DataFrame(boston.target)

# View table
DFboston.head()
```

```python
# Labels of column meanings from the original boston dataset
print(boston.DESCR)
```

```python
# Get some statistics from the dataset
DFboston.describe()
```

```python
#Split the data into X and Y
X = DFboston.drop('Price', axis = 1)
Y = DFboston['Price']
```

```python
#Split the data into 67% training and 33% testing set
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.33 , random_state = 42)
```

```python
# Initialize the linear regression model
reg = linear_model.LinearRegression()
```

```python
# Train the model with our training data
reg.fit(x_train , y_train)
```

```python
#Print the predictions
y_pred = reg.predict(x_test)
print(y_pred)
```

```python
# Check the model performance / accuracy using the Mean Squared Error (MSE)
print(np.mean((y_pred - y_test)**2)) 

# The closer MSE is to 0 the better
```

```python
# Check the model performance / accuracy using the Mean Squared Error (MSE) and sklearn.metrics
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, y_pred))
```
